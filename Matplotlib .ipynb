```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([2,10])
y = np.array([2,10])
font1 = {'family':'serif','color':'black','size':20}
font2 = {'family':'serif','color':'blue','size':15}


plt.subplot(2, 3, 1)
plt.plot(x,y,'o')
plt.title("Title 1")
plt.xlabel("X", fontdict = font2)
plt.ylabel("Y", fontdict = font2)

#plot 2:
y1 = np.array([3,8,3,8])
y2 = np.array([6,2,6,2])

plt.subplot(2, 3, 2)
plt.plot(y1,marker='*',color='orange',linewidth=0.5)
plt.plot(y2,linestyle='dashed',color='blue',linewidth=0.5)

#plot 3:
ypoints = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 3)
plt.plot(ypoints, marker = 'o', ms = 5, mec = 'k', mfc = 'y')

#plot 4:



plt.suptitle("GRAPH")
plt.show()
```
